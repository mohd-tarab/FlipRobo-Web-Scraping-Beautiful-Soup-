# Write a python program to display all the header tags from:-
‘en.wikipedia.org/wiki/Main_Page’.

```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
# finding all headers form wiki home page.

Header_1= soup.find_all('h1')
Header_2=soup.find_all('h2')
Header_3=soup.find_all('h3')
```

```python
# Extracting headers


First_Header=[]

for i in Header_1:
    First_Header.append(i.text.replace('\n',''))
First_Header
```

```python
Second_Header=[]

for i in Header_2:
    Second_Header.append(i.text.replace('\n',''))
Second_Header
```

```python
Third_Header=[]

for i in Header_3:
    Third_Header.append(i.text.replace('\n',''))
Third_Header
```

```python
print(len(First_Header),len(Second_Header),len(Third_Header))
```

```python
import pandas as pd
```

```python
Header1=pd.DataFrame({})
Header1['First_Header']=First_Header

Header2=pd.DataFrame({})
Header2['Second_Header']=Second_Header

Header3=pd.DataFrame({})
Header3['Third_Header']=Third_Header
```

```python
df=pd.concat([Header1,Header2,Header3],axis=1)
```

```python
df=df.fillna('')
df
```
