## Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:
i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
ii) Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
from requests import get
```

```python
# NOW WE WILL SEND GET REQUEST TO THE WEBPAGE SERVER TO GET THE SOURCE CODE OF THE PAGE WHICH IS TO BE SCRAPED BY USING THE REQUEST LIBRARY

ODI = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
ODI
```

```python
#lets see how the webpage scource code downloaded above looks like

ODI.content
```

```python
# Now we will parse this source code by using BS and the parser used here in html.parser as the source code in HTML
odi_soup = BeautifulSoup(ODI.content, 'html.parser')
odi_soup
```

```python
odi_soup.title.text
```

```python
Odi= odi_soup.findAll('div',{'class':'l-sticky-side-nav'})
```

```python
# finding top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
```

```python
odi=Odi[0]
```

```python
# Countries


Name= odi.find_all('span',class_="u-show-phablet") 
Name
```

```python
Country=[]

for i in Name:
    Country.append(i.text.replace('\n',''))
Country
```
