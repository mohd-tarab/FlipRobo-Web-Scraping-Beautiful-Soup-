# Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
from requests import get
```

#### i) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
odi_women_team = requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
odi_women_team
```

```python
soup = BeautifulSoup(odi_women_team.content, 'html.parser')
soup
```

```python
odi_women_team = soup.find_all('span', class_="u-hide-phablet")
```

```python
top_women_teams=[]
for i in odi_women_team:
    top_women_teams.append(i.text.replace('\n',''))
top_women_teams[:10]
```

```python
match = soup.find_all('td',class_='rankings-block__banner--matches')
match
```

```python
women_match=[]
for i in match:
    women_match.append(i.text.replace('\n',''))
women_match
```

```python
match_rows = soup.find_all('td',class_='table-body__cell u-center-text')
```

```python
for i in range(0,len(match_rows),2):
    women_match.append(match_rows[i].text.replace('\n',''))
women_match[:10]
```

```python
odi_womenpoint_row1 = soup.find_all('td',class_='rankings-block__banner--points')
odi_womenpoint_row1
```

```python
womenpoint = []
for i in odi_womenpoint_row1:
    womenpoint.append(i.text.replace('\n',''))
womenpoint
```

```python
odi_womenpoint_otherrows = soup.find_all('td',class_='table-body__cell u-center-text')
```

```python
for i in range(1,len(odi_womenpoint_otherrows),2):
    womenpoint.append(odi_womenpoint_otherrows[i].text)
womenpoint[:10]
```

```python
odi_womenrating_row1 = soup.find_all('td',class_='rankings-block__banner--rating u-text-right')
odi_womenrating_row1
```

```python
womenrating = []
for i in odi_womenrating_row1:
    womenrating.append(i.text.replace('\n','').replace(' ',''))
womenrating
```

```python
odi_womenrating_otherrows = soup.find_all('td',class_='table-body__cell u-text-right rating')
```

```python
for i in odi_womenrating_otherrows:
    womenrating.append(i.text.replace('\n',''))
womenrating[:10]
```

```python
print("\n teams-",len(top_women_teams),"\n matches-",len(women_match),"\n points-",len(womenpoint),"\n rating-",len(womenrating))
```

```python
import pandas as pd
```

```python
ODI_WOMEN_TEAM = pd.DataFrame({})
ODI_WOMEN_TEAM["Teams"] = top_women_teams[:10]
ODI_WOMEN_TEAM["Matches"] = women_match[:10]
ODI_WOMEN_TEAM["Point"] = womenpoint[:10]
ODI_WOMEN_TEAM["Rating"] = womenrating[:10]
ODI_WOMEN_TEAM
```
