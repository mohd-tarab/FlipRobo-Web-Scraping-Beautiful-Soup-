## Write a python program to scrap book name, author name, genre and book review of any 5 books from â€˜www.bookpage.com

```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
from requests import get
```

```python
book1 = requests.get("https://bookpage.com/reviews")
book1
```

```python
book1 = BeautifulSoup(book1.content, 'html.parser')
book1
```

```python
#lets see how the webpage scource code downloaded above looks like

book1.content
```

```python
book1.title.text
```

```python
books= book1.find_all('div',{'class':'span9'})
```

```python
books
```

```python
# extracting book names
```

```python
titles= book1.find_all('h4',class_="italic") 
titles
```

```python
books_titles=[]

for i in titles:
    books_titles.append(i.text.replace('\n',''))
books_titles
```

```python
# extracting book authors
```

```python
author= book1.find_all('p',class_="sans bold") 
author
```

```python
Authors=[]

for i in author:
    Authors.append(i.text.replace('\n',''))
Authors
```

```python
#Extracting Gener
```

```python
gener= book1.find_all('p',class_="genre-links hidden-phone") 
gener
```

```python
Gener=[]

for i in gener:
    Gener.append(i.text.replace('\n',''))
Gener
```

```python
#extracting reviews
```

```python
reviews= book1.find_all('p',class_="excerpt") 
reviews
```

```python
Reviews=[]

for i in reviews:
    Reviews.append(i.text.replace('\n',''))
Reviews
```

```python
import pandas as pd 

BookPage=pd.DataFrame({})
BookPage['Book_titles']=books_titles[:5]
BookPage['Author']=Authors[:5]
BookPage['Gener']=Gener[:5]
BookPage['Reviews']=Reviews[:5]
```
